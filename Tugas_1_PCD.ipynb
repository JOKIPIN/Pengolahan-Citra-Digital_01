**Memodifikasi Pemerograman**


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Akses file citra
image_path = "aku.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Konversi ruang warna BGR(default dalam OpenCV) ke RGB

# Menampilkan Gambar
plt.imshow(image)
plt.show()
```

1. **Pengertian Channel RGB:** Setiap gambar digital terdiri dari tiga warna dasar: Merah (R), Hijau (G), dan Biru (B). Setiap warna diwakili oleh nilai numerik dalam rentang 0 hingga 255.

2. **Mengubah Ukuran Gambar:** Gambar asli seringkali besar, sehingga kita memperkecil ukurannya menjadi 16x16 piksel. Ini memudahkan kita untuk melihat dan menganalisis nilai pixel.

3. **Menampilkan Nilai Pixel:** Setelah gambar diperkecil, kita dapat menampilkan nilai untuk setiap piksel di setiap channel (R, G, dan B) dalam bentuk matriks dua dimensi.
Ini akan menunjukkan bagaimana warna ditentukan di setiap titik gambar.

**Langkah-Langkah melakukan implementasi :**

1. **Membaca dan Mengubah Warna Gambar**: Kita membaca gambar dan mengubah ruang warnanya dari BGR (yang digunakan oleh OpenCV) ke RGB.
2. **Mengubah Ukuran Gambar**: Gambar diubah ukurannya menjadi 16x16 piksel untuk memudahkan analisis.
3. **Menampilkan Gambar**: Gambar yang lebih kecil ditampilkan menggunakan matplotlib.
4. **Mencetak Nilai untuk Setiap Channel**: Nilai setiap channel (R, G, B) untuk setiap piksel ditampilkan dalam format matriks. Ini membantu kita melihat warna apa yang ada di setiap piksel.



```python
image_small = cv2.resize(image, (16, 16))

height, width, channels = image_small.shape

# Tampilkan gambar
plt.figure(figsize=(5, 5))
plt.imshow(image_small)
plt.axis('off')  # Menyembunyikan sumbu
plt.title("Gambar Kecil (16x16)")
plt.show()

# Menampilkan nilai channel
def display_channel_values(channel_name, channel_data):
    print(f"\nNilai channel {channel_name}:")
    for i in range(height):
        print(channel_data[i])

# Menampilkan nilai channel R, G, B
display_channel_values("R", image_small[:, :, 0])
display_channel_values("G", image_small[:, :, 1])
display_channel_values("B", image_small[:, :, 2])
```

# **Representasi Citra RGB**
Dalam ruang warna RGB, setiap warna pada citra dihasilkan dari kombinasi tiga warna dasar cahaya:
- merah (R),
- hijau (G),
- dan biru (B).

**Bagaimana representasi ini bekerja:**

1. *Pixel dan Nilai Intensitas:*

Setiap pixel dalam gambar memiliki tiga nilai, masing-masing mewakili intensitas cahaya untuk warna merah, hijau, dan biru.
- Nilai intensitas ini berkisar dari 0 hingga 255:
0 berarti tidak ada cahaya (warna tersebut tidak muncul).
- 255 berarti cahaya maksimal (warna tersebut muncul dengan intensitas penuh).


2. *Campuran Warna:*

Dengan mencampurkan ketiga warna dasar ini, kita dapat menghasilkan berbagai macam warna:
- Merah murni: (255, 0, 0)
- Hijau murni: (0, 255, 0)
- Biru murni: (0, 0, 255)
- Putih: (255, 255, 255) — semua warna dicampurkan dengan intensitas penuh.
- Hitam: (0, 0, 0) — tidak ada cahaya.

3. *Contoh Campuran:*
- Kuning: (255, 255, 0) — campuran merah dan hijau.
- Cyan: (0, 255, 255) — campuran hijau dan biru.
- Magenta: (255, 0, 255) — campuran merah dan biru.

4.  *Visualisasi:*
Dengan cara ini, citra dapat direpresentasikan sebagai matriks tiga dimensi, di mana setiap dimensi mewakili satu channel warna. Setiap elemen dalam matriks tersebut adalah nilai intensitas warna untuk pixel tertentu.

```python
R, G, B = cv2.split(image) # Pemisahan ketiga channel RGB pada citra ke dalam 3 matriks

blank = np.zeros(image.shape[:2], dtype="uint8")

redImg = cv2.merge([R, blank, blank])
greenImg = cv2.merge([blank, G, blank])
blueImg = cv2.merge([blank, blank, B])

# Plot
plt.figure(figsize=(8, 4))
plt.subplot(2, 3, 1)
plt.imshow(image)
plt.title('Image')

plt.subplot(2, 3, 4)
plt.imshow(redImg)
plt.title('R')

plt.subplot(2, 3, 5)
plt.imshow(greenImg)
plt.title('G')

plt.subplot(2, 3, 6)
plt.imshow(blueImg)
plt.title('B')

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)

plt.show()
```

# Pemisahan Channel RGB ke dalam Warna Hitam-Putih
Pemisahan channel RGB memungkinkan kita untuk melihat intensitas masing-masing warna (merah, hijau, dan biru) secara terpisah. Dengan memetakan setiap channel ke dalam skala hitam-putih, kita bisa lebih mudah memahami seberapa kuat atau lemahnya intensitas warna tersebut dalam gambar.

**Penjelasan Proses:**

1. *Pemisahan Channel:* Setiap gambar RGB memiliki tiga channel: Merah, Hijau, dan Biru.
Ketika kita memisahkan channel ini, kita bisa mengakses nilai intensitas untuk setiap warna.

2. *Pemetaan ke Warna Hitam-Putih:* Nilai intensitas dari setiap channel (0-255) dipetakan ke skala hitam-putih.
- Nilai tinggi (mendekati 255) akan tampil lebih cerah (putih),
- sementara nilai rendah (mendekati 0) akan tampil lebih gelap (hitam).

3. *Interpretasi Hasil:*
Dengan menggunakan visualisasi ini, kita dapat melihat di mana warna tertentu mendominasi dalam gambar:
- Intensitas tinggi pada channel merah, misalnya, akan menghasilkan area cerah pada gambar hitam-putih untuk channel merah.
- Jika sebuah area memiliki intensitas rendah, area tersebut akan terlihat gelap, menunjukkan bahwa warna tersebut tidak terlalu dominan.

```python
# Plot
plt.figure(figsize=(8, 4))
plt.subplot(2, 3, 1)
plt.imshow(image)
plt.title('Image')

plt.subplot(2, 3, 4)
plt.imshow(R, cmap = plt.cm.gray)
plt.title('R (gray)')

plt.subplot(2, 3, 5)
plt.imshow(G, cmap = plt.cm.gray)
plt.title('G (gray)')

plt.subplot(2, 3, 6)
plt.imshow(B, cmap = plt.cm.gray)
plt.title('B (gray)')

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)

plt.show()
```

# Representasi Citra Grayscale
Citra grayscale adalah representasi gambar yang hanya menggunakan satu channel warna, yaitu intensitas abu-abu. Setiap pixel dalam citra grayscale memiliki nilai dari 0 (hitam) hingga 255 (putih), yang merupakan kombinasi dari nilai merah, hijau, dan biru (RGB) dalam citra asli.

Konversi RGB ke Grayscale
Untuk mengonversi citra RGB menjadi grayscale, OpenCV menggunakan persamaan berikut:

$𝑌 = 0.299 × 𝑅 + 0.587 × 𝐺 + 0.114 × 𝐵$
- R: Nilai intensitas warna merah.
- G: Nilai intensitas warna hijau.
- B: Nilai intensitas warna biru.

Persamaan ini memperhitungkan sensitivitas mata manusia terhadap warna. Mata manusia lebih sensitif terhadap cahaya hijau, diikuti oleh merah, dan paling tidak sensitif terhadap biru, sehingga bobot untuk setiap warna berbeda.

[Untuk penjelasan lebih spesifik dapat dilihat disini di bagian cvtColor.](https://docs.opencv.org/2.4/modules/imgproc/doc/miscellaneous_transformations.html#void%20cvtColor%28InputArray%20src,%20OutputArray%20dst,%20int%20code,%20int%20dstCn%29)

```python
# Contoh konversi citra grayscale dari citra RGB

grayImage = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Plot
plt.imshow(grayImage, cmap = plt.cm.gray)
plt.title("Image (Grayscale)")
```

# Representasi Citra Biner
Citra biner adalah representasi gambar yang hanya memiliki dua nilai, yaitu 0 dan 1. Dalam citra biner, setiap pixel ditentukan sebagai:

$0: Mewakili warna hitam (atau area gelap).$
$1: Mewakili warna putih (atau area terang).$

Proses Konversi ke Citra Biner
Salah satu metode umum untuk menghasilkan citra biner adalah dengan menggunakan thresholding.

Berikut adalah langkah-langkahnya:

1. Konversi ke Grayscale: Biasanya, gambar RGB dikonversi terlebih dahulu ke grayscale untuk menyederhanakan proses.

2. Tentukan Nilai Threshold: Sebuah nilai threshold ditentukan. Nilai ini bisa tetap atau ditentukan berdasarkan analisis gambar.

Pemetaan Nilai:

1. Nilai pixel yang lebih tinggi dari threshold akan dipetakan ke 1 (putih).
2. Nilai pixel yang lebih rendah atau sama dengan threshold akan dipetakan ke 0 (hitam).

```python
# Contoh konversi citra biner dari citra grayscale :

thresholdValue = 127  # Nilai threshold yang digunakan
ret, binImage = cv2.threshold(grayImage, thresholdValue, 255, cv2.THRESH_BINARY)  # fungsi pada OpenCV (ret : nilai threshold, binImage : hasil konversi citra)

# Plot
plt.imshow(binImage, cmap = plt.cm.gray)
plt.title("Image (Binary)")
```
